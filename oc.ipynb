```python
import pandas as pd
import matplotlib.pyplot as plt
import flopy.utils.binaryfile as bf
hds = bf.HeadFile('sample' + ".hed") # 获取水头文件
```

```python
idx = (0, 0, 0)
ts = hds.get_ts(idx)
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)
ttl = "Head at cell ({0},{1},{2})".format(idx[0] + 1, idx[1] + 1, idx[2] + 1)
ax.set_title(ttl)
ax.set_xlabel("time/a")
ax.set_ylabel("head/m")
ax.plot(ts[:, 0], ts[:, 1], "bo-")
```

```python
vds = bf.HeadFile('sample' + ".vert_dis",text='Z DISPLACEMENT') # 获取沉降文件
idx = (0, 0, 0)
ts = vds.get_ts(idx)
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(1, 1, 1)
ttl = "subsidence at cell ({0},{1},{2})".format(idx[0] + 1, idx[1] + 1, idx[2] + 1)
ax.set_title(ttl)
ax.set_xlabel("time/a")
ax.set_ylabel("subsidence/m")
ax.plot(ts[:, 0], ts[:, 1], "bo-")
```
