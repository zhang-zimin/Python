```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import mpl
import warnings
warnings.filterwarnings("ignore")

path = r'E:\OneDrive - HHU\西霞院\西霞院基坑震荡实验数据\基坑右岸北井-1.csv'
数据 = pd.read_csv(path,skiprows=11,usecols = ['Time','LEVEL'],encoding = 'gbk')
水位 = 数据['LEVEL']
激发水位 = 水位.max()
激发水位的索引 = int(np.where(水位==激发水位)[0][0])
激发前稳定水位的索引 = 激发水位的索引 - 50
激发前稳定水位 = 水位.iloc[激发前稳定水位的索引]
水位 = 水位.iloc[激发水位的索引-100:激发水位的索引+1700]
x = range(len(水位))
y = 水位
mpl.rcParams['font.sans-serif']=['SimHei'] #黑体
plt.xlabel('时间/s',fontsize=10,color='k')
plt.ylabel('水位/m',fontsize=10,color='k')
ax=plt.axes()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.plot(x,y)
plt.savefig(rf"E:\OneDrive - HHU\西霞院\震荡实验曲线\{path[34:-4]}.png",bbox_inches ='tight',dpi=400)
```
